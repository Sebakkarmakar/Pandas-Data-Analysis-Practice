## Merge two dataframes.

```python
import pandas as pd

df1 = pd.DataFrame({
    "Patient_ID": [101, 102, 103, 104],
    "Name": ["Aman", "Riya", "Karan", "Neha"],
    "Age": [45, 32, 54, 28]
})
df2 = pd.DataFrame({
    "Patient_ID": [101, 102, 103, 105],
    "Disease": ["Liver", "Normal", "Liver", "Normal"],
    "City": ["Delhi", "Mumbai", "Pune", "Chennai"]
})

merge=pd.merge(df1,df2,on='Patient_ID')
print(merge)
```

## Perform inner join.

```python
import pandas as pd

df1 = pd.DataFrame({
    "Patient_ID": [101, 102, 103, 104],
    "Name": ["Aman", "Riya", "Karan", "Neha"],
    "Age": [45, 32, 54, 28]
})
df2 = pd.DataFrame({
    "Patient_ID": [101, 102, 103, 105],
    "Disease": ["Liver", "Normal", "Liver", "Normal"],
    "City": ["Delhi", "Mumbai", "Pune", "Chennai"]
})

merge=pd.merge(df1,df2,on='Patient_ID',how='inner')
print(merge)
```

## Perform left join.

```python
import pandas as pd

df1 = pd.DataFrame({
    "Patient_ID": [101, 102, 103, 104],
    "Name": ["Aman", "Riya", "Karan", "Neha"],
    "Age": [45, 32, 54, 28]
})
df2 = pd.DataFrame({
    "Patient_ID": [101, 102, 103, 105],
    "Disease": ["Liver", "Normal", "Liver", "Normal"],
    "City": ["Delhi", "Mumbai", "Pune", "Chennai"]
})

merge=pd.merge(df1,df2,on='Patient_ID',how='left')
print(merge)
```

## Detect outliers using IQR.

```python
Q1=df['Age'].quantile(0.25)
Q3=df['Age'].quantile(0.75)
IQR=Q3-Q1
lower=Q1-(1.5*IQR)
upper=Q3+(1.5*IQR)
outlier=df[(df['Age']<lower) | (df['Age']>upper)]
print(outlier)
```

## Normalize numeric columns.

```python
numeric_cols = df.select_dtypes(include='number').columns

df[numeric_cols] = (
    df[numeric_cols] - df[numeric_cols].min()
) / (
    df[numeric_cols].max() - df[numeric_cols].min()
)

print(df.head())
```

## Standardize data.

```python
numeric_cols = df.select_dtypes(include='number').columns

df[numeric_cols] = (
    df[numeric_cols] - df[numeric_cols].mean()
) / (
    df[numeric_cols].std()
)

print(df.head())
```
